# DIFFERENT PLOTS USING MATPLOTLIB LIBRARY

## Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

### to draw a plot we have to use a package in matplotlib called pyplot.
 
### PYPLOT - is a module in matplotlib package whichh provides simple functions to add plot elements like lines ,images,text etc to the current axes int the current figure




## simple plot and its function Introduction

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8],"b",linewidth=3)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("first plot")
plt.show()
#matplotlib.pyplot.plot(*args,scale=True,scaley=True,data=None,**kwargs)
#variable length non keyworded arguments
#variable length key word arguments
#data is indexible object ,optional
#scalex,scaley are optional

#when ever we give only one variable values it takes x values as default 0,1,)
```

## LINE PLOT

```python
# parameters order for .plot function is [X],y,[fmt]-format,data=None,**kwargs
#for multiple plot plot([X],y,[fmt],[x2],y2,[fmt2],......,*kwrags)
#formating is like color,marker,linestyle (which cannnot passed as key arguments)

year=[1920,1930,1940,1950,1960,1970,1980,1990,2000]
umemploy=[9.8,12,7.7,6.7,8,8.8,5.4,6.3,7.7]

plt.plot(year,umemploy,marker='o',linestyle="-.")
plt.xlabel("year")
plt.ylabel("unemployment_rate")
plt.title("year vs unemployment_rate")
plt.show()
```

# TYPES OF PLOT

## Line chart or plot 

```python
plt.plot(data['year'])
plt.plot(data['own'])
plt.title("line chart")

#seting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

# to display
plt.show()
```

# Scatter plot

## A scatter plot is a type of graph  displays values for typically two variables for a set of data. It shows how much one variable is affected by another, revealing potential correlations.

```python
#scatter plot year aganist own 
plt.scatter(data['year'],data['own'])

# adding title to the plot 
plt.title("scatter plot")

#x and y labels
plt.xlabel('year')
plt.ylabel('own')

#to display
plt.show()
```

# Bar chart 
## it is a graph that represent the category of data in bar with height and width

```python
plt.bar(data['year'],data['own'])
plt.title("bar chart")


#setting the x and y labels
plt.xlabel('year')
plt.ylabel( 'own')
#to display plot
plt.show()
```

# Histogram 
## Graphical representation tht orginises a group od data points into specified range 

```python
# interval versus frequency that displays large set of data 
#visual representation of data distribution
# x-interval
# y-frequencies

plt.hist(data['income'])
plt.title("histogram")
plt.show()
```

```python
##where ec is edge colour to differecciate between the intervals
plt.hist(data['income'],ec="black")
plt.title("histogram")
plt.show()
```
